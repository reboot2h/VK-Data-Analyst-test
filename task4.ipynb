```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set()
```

```python
df = pd.read_excel('Analytic_test.xlsx', header=1, index_col=0)
df
```

```python
pd.set_option('display.max_columns', None)
df.describe(include='all')
```

```python
df.isna().sum()
```

Исправим названия столбцов: active_Red, Blue_activity_count, Blue_gmv_total  
Заменим NaN на 0

```python
df = df.rename(columns={'active_Red': 'has_Red_id', 'Blue_activity_count': 'Blue_gvm_order_count', 'Blue_gmv_total': 'Blue_gvm_total'})
df = df.fillna(value=0)
```

```python
plt.figure(figsize=(13,5))
ax = sns.barplot(data=(df[['has_Red_id', 'has_Blue_id', 'has_Black_id', 'has_Green_id']]), estimator='sum', errorbar=None, orient='h', palette=['red', 'blue', 'black', 'green'])
ax.set_yticklabels(['Red','Blue','Black', 'Green'])
ax.set(xlabel='Количество пользователей', ylabel='Компания', title='Количество пользователей на компанию')
ax.bar_label(ax.containers[0])
plt.show()
```

Red занимает доминирующую позицию c ~56 тыс. пользователей  
Blue на втором месте с количеством ~14 тыс. пользователей  
У Black и Green всего 1084 и 890 пользователей соответственно

```python
company_combo = {
    "Blue->Red": round(((df.has_Red_id > 0) * (df.has_Blue_id > 0)).sum() / (df.has_Blue_id > 0).sum() * 100),
    "Black->Red": round(((df.has_Red_id > 0) * (df.has_Black_id > 0)).sum() / (df.has_Black_id > 0).sum() * 100),
    "Green->Red": round(((df.has_Red_id > 0) * (df.has_Green_id > 0)).sum() / (df.has_Green_id > 0).sum() * 100),
    "Black->Blue": round(((df.has_Blue_id > 0) * (df.has_Black_id > 0)).sum() / (df.has_Black_id > 0).sum() * 100),
    "Green->Blue": round(((df.has_Blue_id > 0) * (df.has_Green_id > 0)).sum() / (df.has_Green_id > 0).sum() * 100),
    "Green->Black": round(((df.has_Black_id > 0) * (df.has_Green_id > 0)).sum() / (df.has_Green_id > 0).sum() * 100)
}
ax = sns.barplot(y=list(company_combo.keys()), x=list(company_combo.values()), palette=['red', 'red', 'red', 'blue', 'blue', 'black'])
ax.set(xlabel='%', ylabel='Компания A->Компания B', title='Процент пользователей компании A, которые являются пользователями компании B')
ax.bar_label(ax.containers[0])
plt.show()
```

```python
company_combo_2 = {
    "Red": round(((df.has_Red_id > 0) * ((df.has_Blue_id > 0) + (df.has_Black_id > 0) + (df.has_Green_id > 0))).sum() / (df.has_Red_id > 0).sum()  * 100),
    "Blue": round(((df.has_Blue_id > 0) * ((df.has_Red_id > 0) + (df.has_Black_id > 0) + (df.has_Green_id > 0))).sum() / (df.has_Blue_id > 0).sum()  * 100),
    "Black": round(((df.has_Black_id > 0) * ((df.has_Red_id > 0) + (df.has_Blue_id > 0) + (df.has_Green_id > 0))).sum() / (df.has_Black_id > 0).sum()  * 100),
    "Green": round(((df.has_Green_id > 0) * ((df.has_Red_id > 0) + (df.has_Blue_id > 0) + (df.has_Black_id > 0))).sum() / (df.has_Green_id > 0).sum()  * 100)
}
ax = sns.barplot(y=list(company_combo_2.keys()), x=list(company_combo_2.values()), palette=['red', 'blue', 'black', 'green'])
ax.set(xlabel='%', ylabel='Компания', title='Процент пользователей компании, которые являются пользователями другой')
ax.bar_label(ax.containers[0])
plt.show()
```

Лишь малая часть (16%) пользователей Red является пользователями других компаний  
При этом большинство (56-69%) пользователей Blue, Black, Green являются пользователями других компаний, в частности Red, что видно по высоким значениям(45-62%) Blue->Red, Black->Red, Green->Red
